###  Import numpy,pandas,data visualization libraries

```python
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

###  Import the dataset

```python
dataset=pd.read_csv('CAR.csv')
```

```python
dataset
```

###  Understand the dataset

```python
dataset.shape
```

```python
dataset.head()
```

### Visualize the dataset

```python
plt.rcParams["figure.figsize"]=[14,6]
dataset.plot(x='Horsepower',y='Acceleration',style='.',color='blue',markersize=8)
plt.title('Horsepower vs Acceleration')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.grid()
plt.show()
```

###  Corelation

```python
dataset.corr()
```

### Divide the dataset into training and test set

```python
x=dataset.iloc[:,:1].values
```

```python
y=dataset.iloc[:,1:].values
```

```python
x
```

```python
y
```

### Import the function

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

### Prediction

```python
y_pred=lm.predict(x_test)
```

```python
y_pred
```

### Check our prediction

```python
check=pd.DataFrame(x_test,columns=["Horesepower"])
```

```python
check['Acceleration Actual']=y_test
```

```python
check['Acceleration Predicted']=y_pred
```

```python
check
```

### Visualize the regressor line

```python
plt.rcParams["figure.figsize"]=[14,6]
dataset.plot(x='Horsepower',y='Acceleration',style='.',color='blue',markersize=8)
plt.plot(x_test,y_pred,color='red')
plt.title('Horsepower vs Acceleration')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.grid()
plt.show()
```
